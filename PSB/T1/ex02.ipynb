```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
```

```python
t = np.arange(32)
x = np.exp(-0.5*t)
h = [0, 1.5, 0.75, 0.375, 0.1875, 0.09375, 0.046875, 0.0234375]
conv = signal.convolve(x, h, mode='same')
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
ax1.plot(t, x)
ax1.set_title('Signal')
    ax2.plot(t, h)
    ax2.set_title('Digital')
ax3.plot(t, conv)
ax3.set_title('Conv')
```
